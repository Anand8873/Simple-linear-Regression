```python
model.summary()
```

# Square Root Transformation 

```python
model1 = sm.ols('np.sqrt(Salary_Data)~np.sqrt(Years_Experience)', data = df1).fit()
model1.summary()
```

# Log Transformation

```python
model2 = sm.ols('np.log(Salary_Data)~np.log(Years_Experience)', data = df1).fit()
model2.summary()
```

```python
import statsmodels.api as sm
```

```python
qqplot = sm.qqplot(model.resid, line= 'q')
plt.title('Normal Q-Q Plot of Residual')
plt.show()
```

```python
list(np.where(model.resid>10000))
```

```python
df1
```

# Residual plot  Homoscedacity

```python
def standardized(abk):
    return (abk-abk.mean())/abk.std()
```

```python
plt.scatter(standardized(model.fittedvalues),standardized(model.resid))
plt.title('Residual plot')
plt.xlabel('standardized Fitted Values')
plt.ylabel('standardized Residual Values')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
model1_pred_y =np.square(model1.predict(df1['Years_Experience']))
model2_pred_y =np.exp(model2.predict(df1['Years_Experience']))
```

```python
model1_rmse =np.sqrt(mean_squared_error(df1['Salary_Data'], model1_pred_y))
model2_rmse =np.sqrt(mean_squared_error(df1['Salary_Data'], model2_pred_y))
print('model=', np.sqrt(model.mse_resid),'\n' 'model1=', model1_rmse,'\n' 'model2=', model2_rmse)
```

# Model Deletion by Cook's Distance and Leverage value

```python
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
```

```python
plt.stem(np.arange(len(df1)), np.round(c,3))
plt.xlabel('Row Index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
(np.argmax(c), np.max(c))
```
